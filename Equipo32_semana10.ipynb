## 1. Introducción

El objetivo de esta actividad es usar un modelo de lenguaje grande, o **LLM**, como apoyo para identificar y estructurar una idea de proyecto de Inteligencia Artificial aplicable al entorno laboral de cada integrante o a un área de oportunidad relevante. En particular, se busca proponer soluciones que integren temas de **procesamiento de lenguaje natural**, **modelos Transformer**, **generación de texto**, **búsqueda semántica** y **automatización asistida**.

El equipo está conformado por integrantes que trabajan en sectores distintos, por lo que cada uno elaboró una propuesta basada en su propio contexto profesional: **ingeniería de yacimientos (Pemex)**, **marketing y automatización en un despacho de arquitectura** y **soporte técnico de ciberseguridad (Cisco)**. A partir de estas cuatro propuestas individuales, el equipo seleccionó una para desarrollarla a profundidad como *business case*.

La propuesta seleccionada para el desarrollo grupal es un **asistente LLM para triage y troubleshooting de casos de ciberseguridad**. Este contexto es altamente adecuado para NLP porque gran parte del conocimiento operativo se encuentra en texto no estructurado: correos, notas de casos, comandos, logs, documentación de productos, bugs, artículos técnicos y transcripciones de sesiones de troubleshooting. Además, dos integrantes del equipo trabajan directamente en este dominio, lo que aporta experiencia real para desarrollar la solución.

El proyecto seleccionado no busca reemplazar al ingeniero de soporte, sino crear un sistema de apoyo que reduzca tiempo de análisis, mejore la calidad de las respuestas y permita documentar mejor los pasos de diagnóstico, manteniendo siempre revisión humana y controles de seguridad.

### Consideraciones sobre el uso del LLM

El LLM se utiliza como **herramienta de apoyo**, no como fuente única de verdad. En un contexto de ciberseguridad, cualquier respuesta generada debe ser validada por especialistas humanos y contrastada con documentación oficial, procedimientos internos y políticas de seguridad. Además, no se deben ingresar datos reales de clientes, configuraciones sensibles, direcciones IP públicas, credenciales, nombres de empresas o información confidencial en un LLM público sin autorización explícita de la organización.

## 5. Ingeniería de prompts aplicada

En esta sección se documenta el proceso de prompt engineering. El objetivo no fue pedir una respuesta única al LLM, sino hacer varias iteraciones para mejorar progresivamente la propuesta.

### Iteración 1 - Generación inicial de ideas

**Prompt usado:**

> Actúa como consultor de IA aplicada a ciberseguridad. Propón ideas de proyectos basados en NLP y LLM para un ingeniero de soporte técnico que trabaja analizando casos de firewall, logs, configuraciones, correos de clientes y documentación técnica. Incluye problema, datos, modelos, entregable, tecnologías, tiempos, costos y valor para el negocio.

**Síntesis de respuesta obtenida:**

El LLM propuso varias ideas: un asistente de triage para casos, un buscador semántico de documentación, un clasificador de severidad, un generador de respuestas a clientes y un resumidor de logs/casos. La opción de mayor impacto fue el asistente de triage porque combina varias tareas de NLP en un solo flujo.

**Reflexión crítica:**

La primera respuesta fue amplia, pero todavía demasiado general. Fue necesario pedir una versión más aterrizada al contexto de soporte de ciberseguridad y a productos como firewalls, administración centralizada, logs y sesiones de troubleshooting.

### Iteración 2 - Aterrizaje al contexto operativo

**Prompt usado:**

> Toma la idea del asistente de triage y adáptala a un entorno de soporte técnico de ciberseguridad. El sistema debe analizar notas de caso, logs, outputs de CLI y correos de clientes. No debe hacer cambios automáticos en dispositivos. Debe generar hipótesis, comandos sugeridos, documentación relevante y un borrador de actualización al cliente. Propón arquitectura técnica, datos y controles de seguridad.

**Síntesis de respuesta obtenida:**

El LLM recomendó una arquitectura basada en RAG: ingestión de datos, anonimización, extracción de entidades, segmentación de documentos, generación de embeddings, almacenamiento vectorial, recuperación de contexto, generación de respuesta por LLM y validación humana. También recomendó controles como RBAC, auditoría, enmascaramiento de datos sensibles, respuestas con citas y prohibición de acciones autónomas sobre infraestructura.

**Reflexión crítica:**

Esta iteración mejoró el enfoque porque convirtió la idea en una arquitectura de solución. Sin embargo, todavía faltaba definir modelos específicos, métricas de éxito y un plan de implementación medible.

### Iteración 3 - Selección de modelos Transformer y LLM

**Prompt usado:**

> Para el proyecto de asistente LLM de triage técnico, selecciona los modelos de IA adecuados. Distingue entre modelos para generación de texto, embeddings, clasificación, detección de datos sensibles y evaluación. Justifica por qué cada modelo sería útil y menciona alternativas open source y empresariales.

**Síntesis de respuesta obtenida:**

El LLM recomendó usar un modelo generativo grande para razonamiento y redacción, un modelo de embeddings para búsqueda semántica, un clasificador Transformer para categorizar casos, un modelo NER para detectar datos sensibles y métricas automáticas/humanas para evaluar calidad. También propuso una alternativa empresarial con APIs controladas y una alternativa open source con modelos locales.

**Reflexión crítica:**

La respuesta fue útil, pero se ajustó para no depender de un solo proveedor. En una organización con información sensible, se debe considerar una arquitectura flexible: LLM comercial con controles empresariales, modelo local para datos sensibles o flujo híbrido donde solo se envía texto anonimizado al LLM.

### Iteración 4 - Riesgos, ética y seguridad

**Prompt usado:**

> Analiza los riesgos éticos, técnicos y de seguridad de implementar un asistente LLM para soporte de ciberseguridad. Considera alucinaciones, fuga de información, prompt injection, sobreconfianza, privacidad, sesgos, uso de datos de clientes y responsabilidad profesional. Propón mitigaciones concretas.

**Síntesis de respuesta obtenida:**

El LLM identificó riesgos principales: respuestas incorrectas o alucinadas, exposición de información sensible, prompt injection en documentos externos, dependencia excesiva del sistema, uso indebido de datos de clientes y generación de recomendaciones inseguras. Las mitigaciones propuestas incluyeron revisión humana obligatoria, anonimización, filtros de entrada/salida, validación contra documentación oficial, limitación de permisos, registro de auditoría y evaluación periódica.

**Reflexión crítica:**

Esta iteración fue clave porque en ciberseguridad un LLM no debe operar como autoridad autónoma. Su valor está en acelerar análisis, pero la decisión final debe permanecer en el ingeniero responsable.

### Iteración 5 - Business case y plan de implementación

**Prompt usado:**

> Convierte la propuesta en un business case académico. Incluye objetivo, alcance, usuarios, arquitectura, fases, entregables, KPIs, costos aproximados, riesgos, mitigaciones y conclusiones. El proyecto debe poder presentarse como propuesta futura dentro de una organización.

**Síntesis de respuesta obtenida:**

El LLM estructuró la propuesta como un business case con fases de 8 a 10 semanas: definición de alcance, preparación de datos, construcción del índice vectorial, integración con LLM, evaluación con casos históricos anonimizados, piloto controlado y mejora continua. También propuso KPIs como reducción del tiempo de triage, calidad de respuestas, porcentaje de respuestas con fuentes, satisfacción de usuarios y reducción de retrabajo.

**Reflexión crítica:**

La última iteración permitió convertir una idea técnica en una propuesta ejecutiva. El resultado final combina valor operativo, factibilidad técnica y controles de seguridad, por lo que puede utilizarse como base para un proyecto académico o piloto interno.

## 6. Proyecto grupal seleccionado: Asistente LLM para triage y troubleshooting de casos de ciberseguridad

### 6.1 Nombre del proyecto

**CyberSupport LLM Assistant: Asistente inteligente para análisis inicial, documentación y recomendación de próximos pasos en casos de ciberseguridad.**

### 6.2 Problemática

En equipos de soporte técnico de ciberseguridad, una gran parte del tiempo se dedica a leer y conectar información dispersa: descripción del problema, síntomas reportados por el cliente, logs, errores, configuraciones, notas internas, documentación oficial y defectos conocidos. Esta tarea requiere experiencia, memoria técnica y capacidad para identificar patrones.

El problema no es solamente técnico, sino también operativo: dos ingenieros pueden revisar la misma información y llegar a niveles distintos de detalle en el análisis inicial. Además, la presión por responder rápido puede provocar que se omitan pasos de diagnóstico, comandos útiles o referencias documentales.

### 6.3 Objetivo general

Diseñar una propuesta de solución basada en modelos Transformer y LLM que apoye al ingeniero de soporte en el análisis inicial de casos de ciberseguridad, generando hipótesis técnicas, próximos pasos de troubleshooting, documentación relevante y borradores de comunicación al cliente.

### 6.4 Objetivos específicos

1. Recibir texto técnico no estructurado, como logs, notas de caso y descripciones de problemas.
2. Identificar entidades relevantes: producto, versión, módulo, error, interfaz, dirección IP anonimizada, protocolo, puerto, severidad y síntoma.
3. Buscar información relacionada en una base de conocimiento mediante embeddings y recuperación semántica.
4. Generar una respuesta estructurada con hipótesis, comandos sugeridos, documentación relacionada y riesgos.
5. Producir un borrador de actualización al cliente con tono profesional y empático.
6. Mantener revisión humana obligatoria antes de usar cualquier recomendación.

## 7. Alcance del proyecto

### Incluido en el MVP

- Análisis de texto técnico pegado manualmente en el notebook o interfaz.
- Anonimización básica de datos sensibles.
- Clasificación inicial del tipo de caso: conectividad, VPN, autenticación, políticas, NAT, logging, performance, upgrade, HA, integración con identidad, etc.
- Búsqueda semántica en una base de conocimiento controlada.
- Resumen del caso.
- Hipótesis de causa probable.
- Próximos pasos sugeridos.
- Borrador de correo para el cliente.
- Registro de prompts y respuestas para evaluación académica.

### No incluido en el MVP

- Cambios automáticos en firewalls o consolas de administración.
- Acceso directo a sistemas productivos.
- Envío automático de correos.
- Uso de datos reales de clientes sin anonimización.
- Sustitución de la decisión técnica del ingeniero.

### Usuarios objetivo

- Ingenieros de soporte técnico de ciberseguridad.
- Líderes técnicos que revisan casos complejos.
- Nuevos ingenieros durante entrenamiento.
- Equipos de documentación y knowledge management.

```mermaid
flowchart LR
    A[Entrada: notas de caso, logs, correos, outputs CLI] --> B[Preprocesamiento y anonimización]
    B --> C[Extracción de entidades técnicas]
    C --> D[Clasificación del tipo de problema]
    D --> E[Generación de embeddings]
    E --> F[Base vectorial de conocimiento]
    F --> G[Recuperación de contexto relevante]
    G --> H[LLM generativo]
    H --> I[Respuesta estructurada con fuentes]
    I --> J[Revisión humana del ingeniero]
    J --> K[Actualización al cliente / documentación interna]
    J --> L[Retroalimentación para mejora continua]
```

### Descripción del flujo

1. **Entrada:** el ingeniero introduce una descripción del caso, logs o notas.
2. **Anonimización:** se enmascaran IPs, nombres de clientes, correos, dominios, números de serie y otros datos sensibles.
3. **Extracción de entidades:** se identifican productos, versiones, errores, protocolos y síntomas.
4. **Clasificación:** el sistema predice el tipo de problema para orientar la búsqueda.
5. **Búsqueda semántica:** se recuperan documentos relevantes usando embeddings.
6. **Generación:** el LLM produce un análisis estructurado.
7. **Validación humana:** el ingeniero revisa y decide qué usar.
8. **Retroalimentación:** se almacena evaluación de utilidad para mejorar prompts, datos y recuperación.

## 9. Modelos de IA seleccionados y justificación

| Componente | Modelo sugerido | Justificación |
|---|---|---|
| **Generación y razonamiento** | ChatGPT - GPT-5.5 Thinking | Adecuado para estructurar análisis técnico, generar hipótesis, redactar explicaciones y transformar información no estructurada en planes de acción. |
| **Embeddings / búsqueda semántica** | Modelo de embeddings multilingüe o especializado en texto técnico | Permite encontrar documentación relacionada aunque el usuario no use exactamente las mismas palabras que aparecen en los documentos. |
| **Clasificación de tipo de caso** | BERT, RoBERTa o DistilBERT fine-tuned | Útil para clasificar problemas por categoría técnica y apoyar priorización. |
| **NER / detección de datos sensibles** | Modelo NER basado en Transformer o reglas híbridas | Ayuda a identificar IPs, correos, nombres de clientes, dominios, versiones y entidades que deben anonimizarse. |
| **Reranking** | Cross-encoder Transformer | Puede mejorar el orden de documentos recuperados antes de enviarlos al LLM. |
| **Evaluación** | LLM como evaluador + revisión humana | Permite revisar claridad, completitud, consistencia y presencia de fuentes, pero la evaluación final debe ser humana. |

### Alternativas de implementación

- **Opción académica:** notebook en Python, datos simulados/anonimizados, FAISS local y un LLM vía interfaz web.
- **Opción empresarial controlada:** API de LLM aprobada por la organización, almacenamiento cifrado, RBAC, auditoría y monitoreo.
- **Opción local/on-premise:** modelo open source desplegado internamente para evitar exposición de información sensible.

## 16. Ejemplo conceptual de entrada y salida

### Entrada anonimizada

> Cliente reporta que después de un cambio de ISP, el tráfico que antes salía por dos enlaces ahora solamente sale por ISP-1. La interfaz asociada a ISP-2 muestra paquetes de entrada y drops, pero prácticamente no muestra paquetes de salida. Se requiere validar si el problema está relacionado con rutas, NAT, PBR o monitoreo de path.

### Salida esperada del asistente

**Resumen:**  
El caso describe una pérdida de balanceo o encaminamiento por el segundo ISP después de un cambio de conectividad. La señal principal es que ISP-2 recibe tráfico o drops, pero no envía tráfico de salida.

**Hipótesis iniciales:**

1. La ruta por defecto o rutas específicas favorecen únicamente a ISP-1.
2. La configuración de NAT no coincide con la nueva interfaz o nuevo objeto de red.
3. La política de PBR ya no coincide con el tráfico esperado.
4. El monitoreo de path considera caído el enlace ISP-2.
5. Hay asimetría de tráfico o bloqueo upstream.

**Próximos pasos sugeridos:**

- Validar tabla de ruteo.
- Revisar NAT manual y automático asociado a las interfaces de salida.
- Revisar reglas de PBR/path monitoring.
- Confirmar contadores por interfaz.
- Revisar logs de drop y eventos de conexión.
- Comparar configuración antes/después del cambio de ISP.

**Borrador de actualización al cliente:**

> We are currently reviewing the routing, NAT, and policy-based routing behavior after the ISP change. Based on the interface counters, traffic appears to be reaching the second ISP-facing interface, but we do not yet see matching outbound traffic. Our next step is to validate whether the egress selection is being affected by the default route, NAT rules, path monitoring, or PBR configuration. We will continue reviewing the data and provide the next update once we confirm which component is influencing the traffic path.

## 17. Riesgos y mitigaciones

| Riesgo | Impacto | Mitigación |
|---|---|---|
| Alucinaciones del LLM | Recomendaciones incorrectas | Respuestas con fuentes, validación humana y prohibición de cambios automáticos. |
| Exposición de datos sensibles | Riesgo legal/compliance | Anonimización, RBAC, cifrado, uso de ambientes aprobados y minimización de datos. |
| Prompt injection | Manipulación de respuestas | Filtrado de entradas, separación de instrucciones del sistema, validación de documentos y controles OWASP. |
| Sobreconfianza del usuario | Decisiones sin revisión | Capacitación, disclaimers, proceso human-in-the-loop y revisión obligatoria. |
| Datos desactualizados | Diagnóstico incorrecto | Actualizar base de conocimiento y mostrar fecha/fuente de documentos. |
| Sesgo por casos históricos | Repetición de errores previos | Curación de datos, revisión de expertos y evaluación continua. |
| Costos de inferencia | Incremento de gasto | Limitar tokens, usar caché, modelos pequeños para tareas simples y monitoreo de uso. |

## 18. Consideraciones éticas y de gobierno

El proyecto debe seguir principios de uso responsable de IA. En particular:

- No debe procesar información real de clientes sin autorización.
- Debe anonimizar cualquier dato sensible antes de enviarlo a un modelo.
- Debe informar al usuario que la respuesta es una recomendación asistida, no una decisión final.
- Debe registrar prompts, fuentes recuperadas y respuestas para auditoría.
- Debe permitir retroalimentación humana y corrección de errores.
- Debe evitar acciones autónomas en sistemas productivos.
- Debe respetar políticas internas de seguridad, privacidad, retención y clasificación de datos.

En un entorno de ciberseguridad, el LLM debe comportarse como un **copiloto técnico**, no como un operador autónomo.

````python
# Ejemplo simple de anonimización para el MVP conceptual
# En producción, esto debe robustecerse y validarse con seguridad/compliance.

import re

texto = """
Customer ACME Corp reports that host 10.20.30.40 cannot reach 172.16.1.10.
The affected device is FTD-ABC123 and the contact email is admin@example.com.
"""

patrones = {
    r"\b(?:\d{1,3}\.){3}\d{1,3}\b": "[IP_ADDRESS]",
    r"[\w\.-]+@[\w\.-]+": "[EMAIL]",
    r"\bFTD-[A-Z0-9]+\b": "[DEVICE_ID]",
    r"ACME Corp": "[CUSTOMER]",
}

anon = texto
for patron, reemplazo in patrones.items():
    anon = re.sub(patron, reemplazo, anon)

print(anon.strip())
````
